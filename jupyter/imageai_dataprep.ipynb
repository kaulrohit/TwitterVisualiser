```python
import pandas as pd
import shutil
```

```python
df_train = pd.read_csv('cleaned_trinary_training_70.csv')
train_tags = df_train['tags']
train_images = df_train['id']
df_train.head()
```

```python
df_test = pd.read_csv('cleaned_trinary_test_30.csv')
test_tags = df_test['tags']
test_images = df_test['id']
df_test.head()
```

```python
for i,tag in enumerate(train_tags):           
    try:
        print('images/{}'.format(train_images[i]))
        shutil.copy2('../../test/images/{}'.format(train_images[i]),'../../test/crime_dataset/train/{}/'.format(tag))
    except:
        pass

for i,tag in enumerate(test_tags):           
    try:
        shutil.copy2('../../test/images/{}'.format(test_images[i]),'../..//test/crime_dataset/test/{}/'.format(tag))
    except:
        pass
```
